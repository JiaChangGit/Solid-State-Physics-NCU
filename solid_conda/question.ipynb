```python
import matplotlib.pyplot as plt
import numpy as np
%config InlineBackend.figure_formats = ['svg']
```

```python
def get_3d_E(nx, ny, nz, a=1, b=1, c=1):
    """Eigen-energy within a 3-d box potential."""
    return (nx/a)**2 + (ny/b)**2 + (nz/c)**2
```

## Example

```python
# record the energies, where the quantum number, n, is a positive integer from 1 to 200
energies = []
for nx in range(1, 200):
    for ny in range(1, 200):
        for nz in range(1, 200):
            # {a, b, c} are the length of quantum box
            E = get_3d_E(nx, ny, nz, a=1, b=1, c=1)
            energies.append(E)

# convert list to numpy array
energies = np.array(energies)

# keep values that only in a complete sphere
plotRange = (1, 200**2)

# plot the histogram
plt.hist(energies, bins=50, edgecolor='black', range=plotRange)
plt.xlabel('Energy')
plt.ylabel('Number of states')
plt.show()
```

## Question 1: Bulk
(5pts) `num_states`50`for``E``E``n*800``(n+1)*800``num_states[n]``+1``n``1``49``n`50

```python
# record the energies, where the quantum number, n, is a positive integer from 1 to 200
energies = []
for nx in range(1, 200):
    for ny in range(1, 200):
        for nz in range(1, 200):
            # {a, b, c} are the length of quantum box
            E = get_3d_E(nx, ny, nz, a=1, b=1, c=1)
            energies.append(E)

# convert list to numpy array
energies = np.array(energies)

# generate an array with length of 50 to count the number of energy
num_states = np.zeros(50, dtype=int)



for n in range(50):
    for i in range(199**3):
        if ((energies[i] >= (n* 800)) and (energies[i] <= ((n+1)* 800))):
            num_states[n]+= 1

        
xx = np.arange(50)
plt.bar(xx, num_states)
plt.plot(xx, num_states, '.-r')
plt.xlabel('Energy/800')
plt.ylabel('Number of states')
plt.show()
```
