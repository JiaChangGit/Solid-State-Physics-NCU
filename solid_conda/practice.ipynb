## <font color="Blue">Questions 1: single-band in a 3-D BCC lattice<font>
Given the nearest neighbor sites `Rm` of BCC crystal, please modify the Hamiltonian to plot the band structure.
    
<img src="./images/bcc.png" width="20%" align="left">
<img src="./images/Rm_bcc.png" width="70%" align="left">

```python
H = {}

H[(0,0,0)] = np.array([
    [0],
])
H[(1/2,1/2,1/2)] = np.array([
    [-1],
])
H[(-1/2,-1/2,-1/2)] = np.array([
    [-1]
])
H[(-1/2,1/2,1/2)] = np.array([
    [-1],
])
H[(-1/2,-1/2,1/2)] = np.array([
    [-1]
])
H[(1/2,-1/2,1/2)] = np.array([
    [-1],
])
H[(1/2,1/2,-1/2)] = np.array([
    [-1]
])
H[(1/2,-1/2,-1/2)] = np.array([
    [-1]
])
H[(-1/2,1/2,-1/2)] = np.array([
    [-1]
])

sk = [
    ('P', [pi,pi,pi]),
    ('G', [ 0, 0, 0]),
    ('N', [ 0,pi,pi]),
    ('H', [0,0,2*pi]),
    ('G', [ 0, 0 ,0]),
]
k1Arr = np.linspace([pi,pi,pi], [ 0, 0, 0], num=20) # P-G
k2Arr = np.linspace([ 0, 0 ,0], [ 0,pi,pi], num=20) # G-N
k3Arr = np.linspace([ 0,pi,pi], [0,0,2*pi], num=20) # N-H
k4Arr = np.linspace([0,0,2*pi], [ 0, 0 ,0], num=20) # H-G
kArr = np.r_[k1Arr, k2Arr, k3Arr, k4Arr]
```

```python
# solve bands
bands = solve_bands(H, kArr)
print(f'Shape of bands: {bands.shape}')
print(f'Number of calculated k-points: {bands.shape[0]}')
print(f'Number of bands: {bands.shape[1]}')
print()

# make x-axis for plotting
xArr = kpoints_to_xaxis(kArr)
hk_labels = [ s[0] for s in sk ]
hk_points = [ s[1] for s in sk ]
hk_xaxis = kpoints_to_xaxis(hk_points)

# plot bands
for band in bands.T:
    plt.plot(xArr, band, '-o')
for i, _ in enumerate(hk_labels):
    plt.axvline(hk_xaxis[i])
plt.ylabel('Energy (eV)')
plt.xticks(hk_xaxis, hk_labels)
plt.show()
```

2.What is the relation between band structure and DOS in your results? 

The DOS is the density of the lines of the band structure for a specific energy.
Integrating DOS to Fermi level will be equal to total number of occupied.
The number  of orbitals in a band  within the first Brillouin  zone  is equal to  the  number of unit cells  N    in  the  crystal. This  is  much  the  same  as  the  statement  made  in  connection  with  the number  of  lattice  vibrational  modes,  and  is  proved  in  a  like  manner,  by  appealing  to  the boundary conditions. Consider  first  the  one-dimensional  case.  The  allowed  values  of k form  a  uniform  mesh  whose unit spacing is 2/L. The number of states inside the first zone, whose length is 2/a, is therefore equal  to  (2/a)/(2/L)  =  L/a = N,  where N is  the  number  of  unit  cells,  in  agreement  with  the assertion made earlier.
